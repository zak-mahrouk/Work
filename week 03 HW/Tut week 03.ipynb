```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
pen = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
pen
```

```python
sns.kdeplot(data= pen, x='flipper_length_mm', fill=True, color='blue')

plt.title('KDE Plot of Flipper Length in Penguins')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Density')

plt.savefig("kde_flipper_length.png", format="png", dpi=300)

plt.show()
```

```python
mean_val = pen['flipper_length_mm'].mean()
median_val = pen['flipper_length_mm'].median()
min_val = pen['flipper_length_mm'].min()
max_val = pen['flipper_length_mm'].max()
q1 = pen['flipper_length_mm'].quantile(0.25)
q3 = pen['flipper_length_mm'].quantile(0.75)
std_val = pen['flipper_length_mm'].std()

# Create a KDE plot using Seaborn
sns.kdeplot(data=pen, x='flipper_length_mm', fill=True, color='blue')

# Add vertical lines for mean, median, and other statistics
plt.axvline(mean_val, color='green', linestyle='--', label='Mean')
plt.axvline(median_val, color='blue', linestyle='--', label='Median')
plt.axvline(min_val, color='red', linestyle=':', label='Min')
plt.axvline(max_val, color='red', linestyle=':', label='Max')
plt.axvline(q1, color='orange', linestyle='-.', label='25th Percentile (Q1)')
plt.axvline(q3, color='orange', linestyle='-.', label='75th Percentile (Q3)')
plt.axvline(mean_val - 2*std_val, color='purple', linestyle=':', label='Mean - 2*STD')
plt.axvline(mean_val + 2*std_val, color='purple', linestyle=':', label='Mean + 2*STD')

# Add title, labels, and legend using Matplotlib
plt.title('KDE Plot of Flipper Length in Penguins with Key Statistics')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Density')
plt.legend(loc='upper right')

plt.savefig("kde_flipper_length.png", format="png", dpi=300)
# Display the plot
plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Load dataset
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
dataset = pd.read_csv(url)

# List of years to animate over
years = ["1952", "1962", "1967", "1972", "1977", "1982", "1987", "1992", "1997", "2002", "2007"]

# Unique list of continents
continents = dataset["continent"].unique()

# Initialize the figure structure
fig_dict = {
    "data": [],
    "layout": {
        "xaxis": {"range": [30, 85], "title": "Life Expectancy"},
        "yaxis": {"title": "GDP per Capita", "type": "log"},
        "hovermode": "closest",
        "updatemenus": [{
            "buttons": [
                {
                    "args": [None, {"frame": {"duration": 500, "redraw": False},
                                    "fromcurrent": True, "transition": {"duration": 300, "easing": "quadratic-in-out"}}],
                    "label": "Play",
                    "method": "animate"
                },
                {
                    "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                      "mode": "immediate",
                                      "transition": {"duration": 0}}],
                    "label": "Pause",
                    "method": "animate"
                }
            ],
            "direction": "left",
            "pad": {"r": 10, "t": 87},
            "showactive": False,
            "type": "buttons",
            "x": 0.1,
            "xanchor": "right",
            "y": 0,
            "yanchor": "top"
        }]
    },
    "frames": []
}

# Slider configuration
sliders_dict = {
    "active": 0,
    "yanchor": "top",
    "xanchor": "left",
    "currentvalue": {
        "font": {"size": 20},
        "prefix": "Year:",
        "visible": True,
        "xanchor": "right"
    },
    "transition": {"duration": 300, "easing": "cubic-in-out"},
    "pad": {"b": 10, "t": 50},
    "len": 0.9,
    "x": 0.1,
    "y": 0,
    "steps": []
}

# Function to create the data for each continent and year
def create_data(year, continent):
    dataset_by_year = dataset[dataset["year"] == int(year)]
    dataset_by_year_and_cont = dataset_by_year[dataset_by_year["continent"] == continent]
    
    return {
        "x": list(dataset_by_year_and_cont["lifeExp"]),
        "y": list(dataset_by_year_and_cont["gdpPercap"]),
        "mode": "markers",
        "text": list(dataset_by_year_and_cont["country"]),
        "marker": {
            "sizemode": "area",
            "sizeref": 200000,
            "size": list(dataset_by_year_and_cont["pop"])
        },
        "name": continent
    }

# Add initial data for the first year
year = 1952
for continent in continents:
    fig_dict["data"].append(create_data(year, continent))

# Create frames for animation (for each year)
for year in years:
    frame = {"data": [], "name": str(year)}
    
    # Add data for each continent in the given year
    for continent in continents:
        frame["data"].append(create_data(year, continent))
    
    fig_dict["frames"].append(frame)
    
    # Create steps for the slider
    slider_step = {
        "args": [
            [year],
            {"frame": {"duration": 300, "redraw": False},
             "mode": "immediate",
             "transition": {"duration": 300}}
        ],
        "label": year,
        "method": "animate"
    }
    sliders_dict["steps"].append(slider_step)

# Add the slider to the layout
fig_dict["layout"]["sliders"] = [sliders_dict]

# Create and show the figure
fig = go.Figure(fig_dict)
fig.show(renderer="png")
```

### 8. Provide a second version of the figure from the previous problem where you edit the fig = px.scatter() function from the Gapminder code so that x is "percent change", y is "rank", size is "percent", and color="sex", animation_frame is "year", and animation_group and hover_name are "name". Then use size_max=50, range_x=[-0.005,0.005]) and remove the log_x=True and range_y parameters

```python
import pandas as pd
import plotly.express as px

# Load the baby names dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Make identical boy and girl names distinct
bn['name'] = bn['name'] + " " + bn['sex']

# Rank names by percent within each year (lower rank = more popular)
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)

# Sort by name and year
bn = bn.sort_values(['name', 'year'])

# Calculate the percent change from the previous year
bn['percent change'] = bn['percent'].diff()

# Identify new names (first appearance or name change) to prevent calculation across name boundaries
new_name = [True] + list(bn.name[:-1].values != bn.name[1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent']

# Sort the dataset by year
bn = bn.sort_values('year')

# Restrict to "common" names (percent > 0.001)
bn = bn[bn.percent > 0.001]

# Create the scatter plot with the given specifications
fig = px.scatter(
    bn,
    x="percent change",  # X-axis as percent change
    y="rank",            # Y-axis as rank (lower rank = more popular)
    animation_frame="year",  # Animate by year
    animation_group="name",  # Group by name
    size="percent",          # Bubble size as percent
    color="sex",             # Color by sex (gender)
    hover_name="name",       # Hover info as name
    size_max=50,             # Set maximum bubble size
    range_x=[-0.005, 0.005]  # Limit x-axis range
)

# Reverse y-axis to put rank 1 at the top
fig.update_yaxes(autorange='reversed')

# Show the figure as PNG for compatibility with GitHub and MarkUs
fig.show(renderer="png")
```
