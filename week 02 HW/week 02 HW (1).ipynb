### 3. Submit your preferred version of the Monty Hall problem that is verified to be running and working with a final printed output of the code; then, add code comments explaining the purpose of each line of the code

```python
import numpy as np

all_door_options = (1, 2, 3)
initial_choice = 1  # chose what door you want to chose (1,2,3)
win_count = 0 # this reffers to the amount of times the winning door matches your choice
reps = 100000 # this the number of times the for loop will be run

for _ in range(reps):
    # Randomly select the winning door
    winning_door = np.random.choice(all_door_options)
    
    # Host reveals a goat door (not the winning door and not the initial choice)
    possible_reveal_doors = [door for door in all_door_options if door != winning_door and door != initial_choice]
    revealed_goat_door = np.random.choice(possible_reveal_doors)
    
    # The remaining door is the one the player would switch to
    remaining_door = [door for door in all_door_options if door != initial_choice and door != revealed_goat_door][0]
    
    # Player switches to the remaining door
    if remaining_door == winning_door:
        win_count += 1 # increment the winning value (add 1)

# Calculate the win ratio
win_ratio = win_count / reps
win_ratio
```
