### Example in Code:

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

# Step 1: Simulate Data
n = 100  # number of data points
x = np.linspace(0, 10, n)  # predictor variable values
beta0 = 2  # true intercept
beta1 = 0.5  # true slope
sigma = 1.0  # standard deviation of the error term

# Generate normally distributed errors
errors = np.random.normal(0, sigma, n)

# Compute the outcome variable Y
Y = beta0 + beta1 * x + errors

# Create a DataFrame for the simulated data
data = pd.DataFrame({'x': x, 'Y': Y})

# Step 2: Fit the Simple Linear Regression model using statsmodels
fitted_model = smf.ols('Y ~ x', data=data).fit()

# Step 3: Extract coefficients
intercept = fitted_model.params['Intercept']
slope = fitted_model.params['x']

# Step 4: Manually calculate fitted values using the estimated coefficients
fitted_values_manual = intercept + slope * data['x']

# Step 5: Compare with fitted_model.fittedvalues
print("Manual calculation of fitted values (first 5):")
print(fitted_values_manual.head())

print("\nUsing fitted_model.fittedvalues (first 5):")
print(fitted_model.fittedvalues.head())
```

```python
# Import necessary libraries
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import plotly.io as pio

# Set the renderer for plotly to display inline in Jupyter
pio.renderers.default = 'notebook'  # Try 'iframe' or 'colab' if 'notebook' doesn't work

# Define parameters for data generation
x = np.linspace(-3, 3, 30)  # X values ranging from -3 to 3
beta, bias, sigma = 2.5, 5.2, 2.5  # Coefficients for the linear model
errors = stats.norm(0, scale=sigma).rvs(size=len(x))  # Generate random errors
y = beta * x + errors  # Generate Y values

# Create a DataFrame from the data
df = pd.DataFrame({'x': x, 'y': y})

# Fit an OLS model using statsmodels
model = smf.ols("y ~ x", data=df)
fitted_model = model.fit()

# Add predicted values to the DataFrame
df['fitted'] = fitted_model.fittedvalues

# Plot the data points and the fitted line
fig = px.scatter(df, x='x', y='y', title='Ordinary Least Squares Fitting')
fig.add_scatter(x=df['x'], y=df['fitted'], mode='lines', line=dict(color='blue'), name='Fitted Line (OLS)')

# Plot vertical lines showing residuals (errors) from actual points to the fitted line
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]], y=[df['y'][i], df['fitted'][i]], 
                    mode='lines', line=dict(color='orange', dash='dash'), showlegend=False)

# Display the plot
fig.show(renderer="png")   # Adjust the renderer as needed
```

### 10. Let's now consider just the (n=160) long wait times (as specified in the code below), and write code to do the following:

```python
import numpy as np
import statsmodels.formula.api as smf
import plotly.express as px
import seaborn as sns

# Load dataset
old_faithful = sns.load_dataset('geyser')

# Define long wait threshold and filter data
long_wait_limit = 63  # Adjust this as needed
long_wait = old_faithful['waiting'] >= long_wait_limit

# Check if long_wait subset is empty
if old_faithful[long_wait].empty:
    print("Error: The long wait time subset is empty. Check the threshold or dataset.")
else:
    # 1. Bootstrap sampling distribution of slope coefficients
    bootstrap_slope_coefficients = []
    for _ in range(1000):
        sample = old_faithful[long_wait].sample(n=len(old_faithful[long_wait]), replace=True)
        model = smf.ols('duration ~ waiting', data=sample).fit()
        bootstrap_slope_coefficients.append(model.params['waiting'])

    bootstrap_slope_coefficients = np.array(bootstrap_slope_coefficients)

    # 2. Simulated sampling distribution under null hypothesis
    simulated_slope_coefficients = []
    beta_0 = 1.65  # Mean slope from observed data
    sigma = 0.37   # Standard deviation from observed data

    for _ in range(1000):
        simulated_data = old_faithful[long_wait].copy()
        simulated_data['duration'] = beta_0 + sigma * np.random.normal(size=len(simulated_data))
        model = smf.ols('duration ~ waiting', data=simulated_data).fit()
        simulated_slope_coefficients.append(model.params['waiting'])

    simulated_slope_coefficients = np.array(simulated_slope_coefficients)

    # 3. Calculate 95% Confidence Interval if array is not empty
    if len(bootstrap_slope_coefficients) > 0:
        ci_lower, ci_upper = np.percentile(bootstrap_slope_coefficients, [2.5, 97.5])
        print(f"95% CI for bootstrap slope coefficients: [{ci_lower}, {ci_upper}]")
    else:
        print("Error: Bootstrap slope coefficients array is empty.")

    # Calculate and display the original model's p-value
    original_model = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit()
    original_p_value = original_model.pvalues['waiting']
    print(f"Original model p-value: {original_p_value}")

    # Calculate the simulated p-value
    if len(simulated_slope_coefficients) > 0:
        simulated_p_value = np.mean(np.abs(simulated_slope_coefficients) >= np.abs(original_p_value))
        print(f"Simulated p-value under null hypothesis: {simulated_p_value}")
    else:
        print("Error: Simulated slope coefficients array is empty.")
```

The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) hypothesis test; but, we can do "two sample" hypothesis testing using a permuation test, or create a 95% bootstrap confidence interval for the difference in means of the two populations.

### A) and B)

```python
import numpy as np

observed_difference = old_faithful.groupby('kind')['duration'].mean().diff().values[-1]
permuted_differences = []

for _ in range(1000):
    shuffled_labels = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(frac=1, replace=False).values)
    mean_diff = shuffled_labels.groupby('kind_shuffled')['duration'].mean().diff().values[-1]
    permuted_differences.append(mean_diff)

p_value = np.mean(np.abs(permuted_differences) >= np.abs(observed_difference))

bootstrapped_mean_differences = []

for _ in range(1000):
    short_sample = old_faithful[old_faithful['kind'] == 'short']['duration'].sample(frac=1, replace=True)
    long_sample = old_faithful[old_faithful['kind'] == 'long']['duration'].sample(frac=1, replace=True)
    mean_diff = long_sample.mean() - short_sample.mean()
    bootstrapped_mean_differences.append(mean_diff)

# Calculate 95% confidence interval
ci_lower, ci_upper = np.percentile(bootstrapped_mean_differences, [2.5, 97.5])
print(f"95% Confidence Interval: [{ci_lower}, {ci_upper}]")
```

### a) Explain how the sampling approaches work for the two simulations.

##### Test of Permutation:

- This method involves recalculating the difference in means for each shuffle after randomly rearranging the group labels. By generating a distribution of mean differences as though there were no actual group differences, this mimics the null hypothesis.
- By contrasting it with a distribution of differences produced under the null hypothesis, the permutation test evaluates the probability of seeing the true mean difference.
##### Confidence Interval for Bootstrap:

- This method determines the difference in means for each sample by repeatedly sampling with replacement within each group (short and long). This generates a resampled distribution of mean differences.
- A confidence interval around the observed mean difference is provided by the bootstrap method, which represents the range of differences we could anticipate if the study were conducted more than once.

### b) Compare and contrast these two methods with the indicator variable based model approach used in Question 11, explaining how they're similar and different.

##### Comparables:

- The bootstrap confidence interval method and the permutation test both compare the differences between short and lengthy wait durations.
- These techniques seek to ascertain whether there is a statistically significant variation in eruption duration according to wait time classification, much as the indicator variable model.
##### Disparities:

- Indicator Variable Model: This model assumes normally distributed errors and uses a regression approach with an indicator variable to directly estimate the difference in means. To determine whether the difference is significant, it generates a p-value.
- The permutation test is a non-parametric method that does not presuppose a particular residual distribution. Rather, it shuffles group labels to produce a distribution of mean differences.
- The Bootstrap framework Confidence Interval: Neither normality nor any particular distribution are assumed in this method. It provides a non-parametric method of evaluating the range of likely differences by creating a confidence interval via resampling within each group.
