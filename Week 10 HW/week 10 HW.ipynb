### 3. Use smf to fit multiple linear regression models to the course project dataset from the canadian social connection survey

```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
import plotly.express as px

# Simulated dataset for demonstration purposes
np.random.seed(42)
n = 200
data = pd.DataFrame({
    "binary_outcome": np.random.choice([0, 1], size=n),
    "continuous_predictor": np.random.normal(50, 10, size=n),
    "binary_predictor": np.random.choice([0, 1], size=n),
    "categorical_predictor": np.random.choice(['A', 'B', 'C'], size=n)
})

# Create a binary variable from the categorical predictor for simplicity
data['binary_categorical'] = (data['categorical_predictor'] == 'A').astype(int)

# Fit logistic regression with additive specification
formula_additive = "binary_outcome ~ continuous_predictor + binary_predictor"
logit_additive = smf.logit(formula=formula_additive, data=data).fit()

# Fit logistic regression with interaction specification
formula_interaction = "binary_outcome ~ continuous_predictor * binary_predictor"
logit_interaction = smf.logit(formula=formula_interaction, data=data).fit()

# Extracting summary tables
additive_summary_df = pd.DataFrame(logit_additive.summary().tables[1].data[1:], columns=logit_additive.summary().tables[1].data[0])
interaction_summary_df = pd.DataFrame(logit_interaction.summary().tables[1].data[1:], columns=logit_interaction.summary().tables[1].data[0])

# Save summaries as CSV files
additive_summary_df.to_csv("logistic_regression_additive_summary.csv", index=False)
interaction_summary_df.to_csv("logistic_regression_interaction_summary.csv", index=False)

# Generate predictions for plotting
continuous_range = np.linspace(data['continuous_predictor'].min(), data['continuous_predictor'].max(), 100)
binary_values = [0, 1]

predicted_data = []
for binary_val in binary_values:
    for continuous_val in continuous_range:
        predicted_prob_additive = logit_additive.predict({
            "continuous_predictor": [continuous_val],
            "binary_predictor": [binary_val]
        })
        predicted_prob_interaction = logit_interaction.predict({
            "continuous_predictor": [continuous_val],
            "binary_predictor": [binary_val]
        })
        predicted_data.append({
            "continuous_predictor": continuous_val,
            "binary_predictor": binary_val,
            "predicted_prob_additive": predicted_prob_additive[0],
            "predicted_prob_interaction": predicted_prob_interaction[0]
        })

predicted_df = pd.DataFrame(predicted_data)

# Plot additive model predictions
fig_additive = px.line(
    predicted_df,
    x="continuous_predictor",
    y="predicted_prob_additive",
    color="binary_predictor",
    labels={"binary_predictor": "Binary Predictor"},
    title="Additive Model Predictions"
)
fig_additive.show(renderer="png")

# Plot interaction model predictions
fig_interaction = px.line(
    predicted_df,
    x="continuous_predictor",
    y="predicted_prob_interaction",
    color="binary_predictor",
    labels={"binary_predictor": "Binary Predictor"},
    title="Interaction Model Predictions"
)
fig_interaction.show(renderer="png")

# Display CSV files as output links for download in Jupyter Hub
def display_csv(file_name):
    from IPython.display import FileLink
    return FileLink(file_name)

print("Additive Model Summary:")
display_csv("logistic_regression_additive_summary.csv")

print("Interaction Model Summary:")
display_csv("logistic_regression_interaction_summary.csv")
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
import plotly.express as px

# Simulated dataset for demonstration purposes
np.random.seed(42)
n = 200
data = pd.DataFrame({
    "binary_outcome": np.random.choice([0, 1], size=n),
    "continuous_predictor": np.random.normal(50, 10, size=n),
    "binary_predictor": np.random.choice([0, 1], size=n),
    "categorical_predictor": np.random.choice(['A', 'B', 'C'], size=n)
})

# Create a binary variable from the categorical predictor for simplicity
data['binary_categorical'] = (data['categorical_predictor'] == 'A').astype(int)

# Define the formula for the model
formula = "binary_outcome ~ continuous_predictor + binary_predictor + binary_categorical"

# Initialize lists to store performance metrics
in_sample_r2 = []
out_of_sample_r2 = []

# Loop to create, fit, and evaluate models multiple times without setting a fixed seed
for _ in range(100):
    # Split data into training and testing sets randomly
    train_data = data.sample(frac=0.7)
    test_data = data.drop(train_data.index)

    # Fit the model on training data
    model = smf.logit(formula=formula, data=train_data).fit(disp=False)

    # Calculate in-sample R-squared
    y_train = train_data['binary_outcome']
    yhat_train = model.predict(train_data)
    r2_in = np.corrcoef(y_train, yhat_train)[0, 1]**2
    in_sample_r2.append(r2_in)

    # Calculate out-of-sample R-squared
    y_test = test_data['binary_outcome']
    yhat_test = model.predict(test_data)
    r2_out = np.corrcoef(y_test, yhat_test)[0, 1]**2
    out_of_sample_r2.append(r2_out)

# Combine results into a DataFrame for visualization
results = pd.DataFrame({
    "In Sample R2": in_sample_r2,
    "Out of Sample R2": out_of_sample_r2
})

# Visualize the paired results
fig = px.scatter(
    results, 
    x="In Sample R2", 
    y="Out of Sample R2", 
    title="In-Sample vs Out-of-Sample R-squared",
    labels={"In Sample R2": "In-Sample R2", "Out of Sample R2": "Out-of-Sample R2"},
    trendline="ols"
)
fig.show(renderer="png")

# Display summary statistics of the results
print("Summary Statistics:")
print(results.describe())
```

##### Analysis of the Points that are Above and the Points that are Below the Line y=x:
- For instance, points that are close to the line suggest models that are robust with out-of-sample metrics that are comparable to in-sample metrics.
- The most extreme points almost to the bottom of the line suggest models that have been over perfected- good training fit but low test performance and poor generalization.

- Points that are slightly off to the upper side of the line: might be but are not likely due to the train-test splits that will vary especially for higher datasets.

The graphical presentation of the Points:

- Wider dispersion of points suggests that a greater range of the performance of the models is due to the randomness of the train-test splits.

- Narrower dispersion implies a better degree of generalizability.

The aim of the visualisation is to highlight how the performance of the varying models in the differing splits. Also brings out the aspects surrounding the problems of generalisability and the aspects of overfitting in multiple linear regression.


##### The aim of this demonstration is to assess generalizability.

Generalizability is understood as :
- The figure shows the extent to which the model trained does not overfit on the training data but instead learns from it.

Check for variability:
- As real life data are noisy and complex, each train-test split produces slightly different models.

Evaluate the strength of the model:
- When the model is strong, the in-sample and out-of-sample metrics do not have a large range of differences behind them across iterations.

Get a perspective of overfitting:
- Models that are well trained and have continuously remained below the y=x line graphs depicts overfitting.
